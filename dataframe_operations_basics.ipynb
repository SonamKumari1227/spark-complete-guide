```python
"""
File: dataframe_operations_basics.py

Concepts Covered:
1. Create DataFrame
2. Select Columns
3. Filter Rows
4. Add Derived Columns
5. Rename Columns
6. Drop Columns
7. Sort Data
8. Distinct Values
9. Group By Aggregations
10. Handling Nulls
"""

from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder \
    .appName("DataFrame Basics") \
    .getOrCreate()
```

```python
data = [
    (1, "Alice", "Sales", 50000),
    (2, "Bob", "IT", 70000),
    (3, "Charlie", "Sales", 60000),
    (4, "David", "HR", None)
]

columns = ["id", "name", "department", "salary"]

df = spark.createDataFrame(data, columns)

# Select
df.select("name", "salary").show()
```

```python
# Filter
df.filter(col("salary") > 55000).show()
```

```python
# Add Column
df = df.withColumn(
    "bonus",
    col("salary") * 0.10
)
df.show()
```

```python

# Rename Column
df = df.withColumnRenamed(
    "department",
    "dept"
)

df.show()
```

```python
# Drop Column
df.drop("bonus").show()
```

```python

# Distinct
df.select("dept").distinct().show()
```

```python

# Sort
df.orderBy(
    col("salary").desc()
).show()
```

```python
# Handle Nulls
df.fillna({"salary": 0}).show()

```

```python
# Aggregation
df.groupBy("dept").agg(
    avg("salary").alias("avg_salary"),
    max("salary").alias("max_salary")
).show()

```

```python
spark.stop()
```
